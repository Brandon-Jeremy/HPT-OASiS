```python
import requests
from bs4 import BeautifulSoup
import html5lib
```

```python
URL = "https://edition.cnn.com/markets/stocks/AAPL" 
r = requests.get(URL)
  
soup = BeautifulSoup(r.content, 'html5lib')

news = soup.find_all('span', class_='markets-company-news__item--title')
news
```

```python
titles = []
for title in news:
    titles.append(title.contents[0])

import pandas as pd

df = pd.DataFrame(titles, columns=['News Titles'])
df
```

```python
from textblob import TextBlob

df['Sentiment Polarity'] = df['News Titles'].apply(lambda x: TextBlob(x).sentiment.polarity)
df
```

```python
df.to_excel('cnn_scrape_polarity.xlsx')
```
