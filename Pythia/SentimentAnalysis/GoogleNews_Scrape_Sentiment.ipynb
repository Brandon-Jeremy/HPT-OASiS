```python
def get_article(text):
  dt = []
  gn = GoogleNews(lang = 'en', country=' US')
  search = gn.search(text)

  for article in search['entries']:
    # print("{} : {}".format(article['published'], article['title']))
    dt.append((article['published'], article['title']))
  return dt

data = get_article("AAPL Stock")
print(data)
```

```python
df = pd.DataFrame(data, columns=['Publication', 'Title'])
```

```python
df.head()
```

```python
from textblob import TextBlob
import numpy as np

def polarity_indicator(text):
  return TextBlob(text).sentiment.polarity

df = df.sort_values(ascending=False, by='Publication')
df['Sentiment'] = df['Title']
df['Sentiment'] = df['Sentiment'].apply(polarity_indicator)

df['Sentiment Class'] = np.where(df['Sentiment']>=.5,'Strong Positive', np.where(df['Sentiment']>0,'Positive', np.where(df['Sentiment']<=-.5,'Strong Negative',np.where(df['Sentiment']<0,'Negative','Neutral'))))
df['Sentiment Class'].value_counts()
```

```python
df['Sentiment Class'].value_counts(normalize=True)
```

```python
df['Date'] = pd.to_datetime(df['Publication'])
df['Date'] =df['Date'].dt.date
```

```python
df = df.sort_values(by='Date', ascending=True)
df
```

```python
df.to_excel('output_file.xlsx')
from google.colab import files
files.download('output_file.xlsx')
```
